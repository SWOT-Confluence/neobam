```r
batch_neobam=function(input_file){
    library(dplyr)
    library(hydroGOF)
    library(ggplot2)
    input_file=input_file[[1]] #list structure

stan_file='/nas/cee-water/cjgleason/colin/neobam/neobam/neobam_height2.stan'
source('/nas/cee-water/cjgleason/colin/neobam/neobam/process.R')
source('/nas/cee-water/cjgleason/colin/neobam/neobam/neobam_functions.R')
source('/nas/cee-water/cjgleason/colin/neobam/neobam/make_prior_Q.R')
this_reach_id=substr(input_file,54,64)
gauge_df=readRDS('/nas/cee-water/cjgleason/colin/analyze confluence runs/devset_gauges.RDS')
    this_gauge=filter(gauge_df,reach_id==this_reach_id)%>%
        mutate(date=gauge_time)

   # suppressMessages({
    neobam_output= process_data(readRDS(input_file), stan_file) 
       # })
    
    if(neobam_output$diditrun==TRUE){
     priorQ=make_prior_Q(input_file)
    

neobam_results=data.frame(recon_Q=neobam_output$recon_Q,
                          neobam_time=neobam_output$neobam_time,
                          recon_Q_sd=neobam_output$recon_sd,
                          posterior_Q=neobam_output$posterior_Q,
                          posterior_sd=neobam_output$posterior_Q_sd) %>%
        filter(!is.na(neobam_time))%>%
        filter(!is.na(posterior_Q))%>%
        #change to day to join
        mutate(date=as.Date(neobam_time))%>%
        #join the gauge dataframe
        left_join(this_gauge,by='date')%>%
        left_join(priorQ,by=c('date','reach_id'))%>%
            filter(!is.na(gauge_Q))%>%
     mutate(reconNSE=NSE(recon_Q,gauge_Q))%>%
            mutate(reconspearman=cor(recon_Q,gauge_Q, method =  "spearman",use="complete.obs"))%>%
            mutate(reconSDRR=sd((recon_Q-gauge_Q)/gauge_Q,na.rm=TRUE))%>%
            # mutate(postsigE= ((recon_Q-gauge_Q)/gauge_Q) -   (mean(recon_Q-gauge_Q,na.rm=TRUE)/gauge_Q)   )%>%
            mutate(nt=nrow(.))%>%
            mutate(reconnbias=sum(recon_Q-gauge_Q,na.rm=TRUE)/ (nt*mean(gauge_Q,na.rm=TRUE)))%>%
            mutate(reconabs_bias=abs(reconnbias))%>%
    
     mutate(postNSE=NSE(posterior_Q,gauge_Q))%>%
            mutate(postspearman=cor(posterior_Q,gauge_Q, method =  "spearman",use="complete.obs"))%>%
            mutate(postSDRR=sd((posterior_Q-gauge_Q)/gauge_Q,na.rm=TRUE))%>%
            # mutate(postsigE= ((posterior_Q-gauge_Q)/gauge_Q) -   (mean(posterior_Q-gauge_Q,na.rm=TRUE)/gauge_Q)   )%>%
            mutate(nt=nrow(.))%>%
            mutate(postnbias=sum(posterior_Q-gauge_Q,na.rm=TRUE)/ (nt*mean(gauge_Q,na.rm=TRUE)))%>%
            mutate(postabs_bias=abs(postnbias))%>%
    
     mutate(priorNSE=NSE(priorQ,gauge_Q))%>%
            mutate(priorspearman=cor(priorQ,gauge_Q, method =  "spearman",use="complete.obs"))%>%
            mutate(priorSDRR=sd((priorQ-gauge_Q)/gauge_Q,na.rm=TRUE))%>%
            # mutate(priorsigE= ((priorQ-gauge_Q)/gauge_Q) -   (mean(priorQ-gauge_Q,na.rm=TRUE)/gauge_Q)   )%>%
            mutate(nt=nrow(.))%>%
            mutate(priornbias=sum(priorQ-gauge_Q,na.rm=TRUE)/ (nt*mean(gauge_Q,na.rm=TRUE)))%>%
            mutate(priorabs_bias=abs(priornbias))%>%
    
            select(-gauge_time)

p5=ggplot(neobam_results)+
     geom_ribbon(aes(x=date,ymin=priorQ-priorsd,
                     ymax=priorQ+priorsd),fill='red',alpha=0.2)+
     geom_line(aes(x=date,y=priorQ),col='red')+
     geom_line(aes(x=date,y=gauge_Q),col='black',lwd=1.5)+
  # geom_line(aes(x=date,y=posterior_Q),col='blue')+
  #       geom_ribbon(aes(x=date,ymin=posterior_Q-posterior_sd,
  #                    ymax=posterior_Q+posterior_sd),fill='blue',alpha=0.2)+
  geom_line(aes(x=date,y=recon_Q),col='blue')+
        geom_ribbon(aes(x=date,ymin=recon_Q-recon_Q_sd,
                     ymax=recon_Q+recon_Q_sd),fill='blue',alpha=0.2)+


annotate('text',label=paste0('NSE:prior= ',round(neobam_results$priorNSE[1],digits=2),
                             # ' posterior= ',round(neobam_results$postNSE[1],digits=2),
                              'recon= ',round(neobam_results$reconNSE[1],digits=2)),
         x=min(neobam_results$date),
         y=max(neobam_results$posterior_Q+neobam_results$posterior_sd,na.rm=TRUE),
         hjust=0) +

annotate('text',label=paste0('spearman:prior',round(neobam_results$priorspearman[1],digits=2),
                             # ' posterior',round(neobam_results$postspearman[1],digits=2),
                              'recon',round(neobam_results$reconspearman[1],digits=2)),
         x=min(neobam_results$date),
         y=0.95*max(neobam_results$posterior_Q+neobam_results$posterior_sd,na.rm=TRUE),
         hjust=0) +

annotate('text',label=paste0('SDRR:prior',round(neobam_results$priorSDRR[1],digits=2),
                             # ' posterior',round(neobam_results$postSDRR[1],digits=2),
                              'recon',round(neobam_results$reconSDRR[1],digits=2)),
         x=min(neobam_results$date),
         y=0.9*max(neobam_results$posterior_Q+neobam_results$posterior_sd,na.rm=TRUE),
         hjust=0) +
         
annotate('text',label=paste0('absbias:prior',round(neobam_results$priorabs_bias[1],digits=2),
                             # ' posterior',round(neobam_results$postabs_bias[1],digits=2),
                              'recon',round(neobam_results$reconabs_bias[1],digits=2)),
         x=min(neobam_results$date),
         y=0.85*max(neobam_results$posterior_Q+neobam_results$posterior_sd,na.rm=TRUE),
         hjust=0) 
    
    output=list('neobam_results'=neobam_results,'plotobj'=p5)

saveRDS(output,paste0('/nas/cee-water/cjgleason/colin/neobam20_files/output_MCMC/neobam_output_',this_reach_id,'.rds'))
    # return(output)
    }
                              
}
```

```r
batch_opt=function(input_file){
    source('/nas/cee-water/cjgleason/colin/neobam/neobam/optimizeneobam.R')
library(dplyr)
library(hydroGOF)
    
priorQ=make_prior_Q(input_file)
    
    # print('priorQ')
    # print(priorQ)
posteriorQ=optimizeneobam(input_file)
    # print('posteriorQ')
    # print(posteriorQ)
    
  
this_reach_id=substr(input_file,54,64)
    # print('this reach id')
    # print(this_reach_id)
    
this_gauge=filter(gauge_df,reach_id==this_reach_id)%>%
        mutate(date=as.Date(gauge_time))

test= left_join(priorQ,posteriorQ,by=c('date','reach_id'))%>%
        left_join(this_gauge,by=c('date','reach_id'))
    
    # print(test)
    
    if(sum(!is.na(test$posteriorQ)<2)){return('bonk')}
    if(sum(!is.na(test$priorQ)<2)){return('bonk')}

   priorNSE=NSE(test$priorQ,test$gauge_Q)
   priorSpearman=cor(test$priorQ,test$gauge_Q, method =  "spearman",use="complete.obs")
   # priorSigE1= ((test$priorQ-test$gauge_Q)/test$gauge_Q) -   (mean(test$priorQ-test$gauge_Q,na.rm=TRUE)/test$gauge_Q)
   # priorSigE=quantile(abs(priorSigE1), probs =0.68,na.rm=TRUE)
   priorSDRR=sd((test$priorQ-test$gauge_Q)/test$gauge_Q,na.rm=TRUE)
   priorAbsbias=abs(sum(test$priorQ-test$gauge_Q,na.rm=TRUE)/ (length(test$gauge_Q)*mean(test$gauge_Q,na.rm=TRUE)))
     
   posteriorNSE=NSE(test$posteriorQ,test$gauge_Q)
   posteriorSpearman=cor(test$posteriorQ,test$gauge_Q, method =  "spearman",use="complete.obs")
   posteriorSDRR=sd((test$posteriorQ-test$gauge_Q)/test$gauge_Q,na.rm=TRUE)
   # posteriorSigE1= ((test$posteriorQ-test$gauge_Q)/test$gauge_Q) -   (mean(test$posteriorQ-test$gauge_Q,na.rm=TRUE)/test$gauge_Q)
   # posteriorSigE=quantile(abs(posteriorSigE1), probs =0.68,na.rm=TRUE)
   posteriorAbsbias=abs(sum(test$posteriorQ-test$gauge_Q,na.rm=TRUE)/ (length(test$gauge_Q)*mean(test$gauge_Q,na.rm=TRUE)))

stat_df=data.frame(priorNSE=priorNSE,
                   priorSpearman=priorSpearman,
                   priorSDRR=priorSDRR,
                   priorAbsbias=priorAbsbias,
                   posteriorNSE=posteriorNSE,
                   posteriorSpearman=posteriorSpearman,
                   posteriorSDRR=posteriorSDRR,
                   posteriorAbsbias=posteriorAbsbias)
    
    p5=ggplot(test)+
     geom_ribbon(aes(x=date,ymin=priorQ-priorsd,
                     ymax=priorQ+priorsd),fill='red',alpha=0.2)+
     geom_line(aes(x=date,y=priorQ),col='red')+
     geom_line(aes(x=date,y=gauge_Q),col='black',lwd=1.5)+
  geom_line(aes(x=date,y=posteriorQ),col='blue')+
        geom_ribbon(aes(x=date,ymin=posteriorQ-posteriorsd,
                     ymax=posteriorQ+posteriorsd),fill='blue',alpha=0.2)+

annotate('text',label=paste0('NSE:prior= ',round(priorNSE,digits=2),
                             ' posterior= ',round(posteriorNSE,digits=2)),
         x=min(test$date),
         y=7500,
         hjust=0) +
annotate('text',label=paste0('spearman:prior= ',round(priorSpearman,digits=2),
                             ' posterior= ',round(posteriorSpearman,digits=2)),
         x=min(test$date),
         y=7200,
         hjust=0) +
annotate('text',label=paste0('SDRR:prior= ',round(priorSDRR,digits=2),
                             ' posterior= ',round(posteriorSDRR,digits=2)),
         x=min(test$date),
         y=6900,
         hjust=0) +
         
annotate('text',label=paste0('absbias:prior= ',round(priorAbsbias,digits=2),
                             ' posterior= ',round(posteriorAbsbias,digits=2)),
         x=min(test$date),
         y=6600,
         hjust=0) 
   
   output=list('Qdf'=test,'statdf'=stat_df,'plotobj'=p5)
   saveRDS( output, paste0( '/nas/cee-water/cjgleason/colin/neobam20_files/output/' ,
                           this_reach_id,'_optneobam.rds'))

}


```
